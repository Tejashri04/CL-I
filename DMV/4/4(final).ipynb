```python
df = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'TransactionDate': ['2023-01-01', '2023-03-05', '2023-06-15', '2024-01-10', '2024-02-01'],
    'PropertyType': ['Apartment', 'House', 'House', 'Apartment', 'House'],
    'Location': ['New York', 'San Francisco', 'New York', 'Chicago', 'New York'],
    'Price': [1500, 2500, 2000, 3000, 2200]
})
```

```python
# Convert 'TransactionDate' to datetime type for easier filtering
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
```

```python
# 2. Filter by property type (e.g., only 'House' type)
filtered_by_type = df[df['PropertyType'] == 'House']
filtered_by_type

# 3. Filter by location (e.g., only properties in 'New York')
filtered_by_location = df[df['Location'] == 'New York']
filtered_by_location
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['PropertyType_encoded'] = label_encoder.fit_transform(df['PropertyType'])
```

```python
#Grouping by 'location' (neighborhood) and calculating the average price
avg_price_by_location = df.groupby('Location')['Price'].mean().reset_index()
avg_price_by_location 
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
uppet_bound = Q3 + 1.5 * IQR
```

```python
df_no_outliers_price = df[(df['Price'] >= lower_bound) & (df['Price'] <= uppet_bound)]
```

```python
df_no_outliers_price
```

```python
from scipy.stats import zscore
# 2. Identifying Outliers Using Z-score
df['z_score_price'] = zscore(df['Price'])
df_no_outliers_zscore = df[df['z_score_price'].abs() <= 3]

# Final DataFrame without outliers
final_df = df_no_outliers_price[df_no_outliers_price.index.isin(df_no_outliers_zscore.index)]
```
