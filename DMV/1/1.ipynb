```python
#Splitting Columns
if 'location' in unified.columns:
    unified_df[['number','name']] = unified_df['location'].str.split('',1,expand = True)
```

```python
unified_df.columns = unified_df.columns.str.lower().str.strip()
```

```python
unified_df[['number', 'name']] = unified_df['location'].str.split(' ', 1, expand=True)
```

```python
# Derive a new column 'total_sales' if 'quantity' and 'unit_price' exist
if 'quantity' in unified_df.columns and 'price' in unified_df.columns:
    unified_df['total_sales'] = unified_df['quantity'] * unified_df['price']
```

```python
unified_df['total_sales']
```

```python
# Convert order_date to datetime format
unified_df['invoice date'] = pd.to_datetime(unified_df['invoice date'], errors='coerce')

# Extract year, month, and day
unified_df['year'] = unified_df['invoice date'].dt.year
unified_df['month'] = unified_df['invoice date'].dt.month
unified_df['day'] = unified_df['invoice date'].dt.day
```

```python
unified_df['year'] 
```

```python
# Get basic descriptive statistics
print(unified_df.describe())
```

```python
total_sales = unified_df['total_sales'].sum()
print(f"Total Sales: ${total_sales:.2f}")
```

```python
aov = unified_df['total_sales'].mean()
print(f"Average order value : ${aov : .2f}")
```

```python
# Assuming there is a column named 'category'
sales_by_category = unified_df.groupby('category')['total_sales'].sum().sort_values(ascending=False)
sales_by_category
```

```python
# Aggregate total sales by month and year
monthly_sales = unified_df.groupby(['year', 'month'])['total_sales'].sum().reset_index()
print(monthly_sales)
```

```python
#Bar Plot for Sales by Product Category
plt.figure(figsize=(10,6))
sns.barplot(x=sales_by_category.index,y=sales_by_category.values)
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
#plt.xticks(rotation= 45)
plt.show()
```

```python
#Pie Chart for Product Category Distribution
plt.figure(figsize =(8,8))
unified_df['category'].value_counts().plot.pie(autopct='%1.1f%%',colors=sns.color_palette('Set3'))
plt.title('Product Category Distribution')
plt.ylabel('')
plt.show()
```

```python
#Box Plot for Total Sales Distribution

plt.figure(figsize=(10, 6))
sns.boxplot(data=unified_df, x='category', y='total_sales')
plt.title('Total Sales Distribution by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
#Line Plot for Monthly Sales Trend
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='month', y='total_sales', hue='year', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(title='Year')
plt.show()
```
